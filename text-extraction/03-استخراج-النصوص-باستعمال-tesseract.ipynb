```python
image_path = '../output/images/aws-cloud-adoption-framework-pg-06.png'
```

```python
ls {image_path}
```

```python
import pytesseract
from PIL import Image

# Set the path to the Tesseract executable (might not be necessary if it's in your PATH)
# pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

# Load the image

img = Image.open(image_path)

# Perform OCR with Arabic language
text = pytesseract.image_to_string(img, lang='ara')

# Print the extracted text
print(text)

# Optionally, save the text to a file
with open('../output/tesseract-ara-output.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```
